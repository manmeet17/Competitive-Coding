```python
import requests
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
url="https://codeforces.com/api/problemset.problems?tags=implementation"
```

```python
res=requests.get(url=url)
```

```python
res
```

```python
results=res.json()
```

```python
import seaborn as sns
import pandas as pd
```

```python
problems=results['result']['problems']
problems
```

```python
import json
```

```python
df=pd.DataFrame(problems)
df.set_index('contestId', inplace=True)
```

```python
df.head()
```

```python
df.points.value_counts()
sns.set(rc={'figure.figsize':(15,10)})
```

```python
g=sns.countplot(x=df.rating)
plt.xticks(rotation=90)
```

```python
meetu="https://codeforces.com/api/user.status?handle=meetutarun16"
resp=requests.get(meetu).json()
```

```python
resp=resp['result']
```

```python
import pprint
len(resp)
```

```python
solved=[]
for i in resp:
    if i['verdict']=="OK":
        solved.append(i)
```

```python
solved=pd.DataFrame(solved)
```

```python
solved.head()
```

```python
solved.shape
```

```python
problemsSol=list(solved['problem'])
```

```python
problemsSol=pd.DataFrame(problemsSol)
```

```python
sns.countplot(x=problemsSol['rating'])
```
